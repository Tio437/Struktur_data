```python
class Calculator:
    def __init__(self, initial_value):
        self.stack = [initial_value]

    def add(self, value):
        result = self.stack[-1] + value
        self.stack.append(result)
        return result

    def subtract(self, value):
        result = self.stack[-1] - value
        self.stack.append(result)
        return result

    def multiply(self, value):
        result = self.stack[-1] * value
        self.stack.append(result)
        return result

    def divide(self, value):
        result = self.stack[-1] / value
        self.stack.append(result)
        return result

    def undo(self):
        if len(self.stack) > 1:
            self.stack.pop()
            return self.stack[-1]
        return "Tidak ada operasi yang bisa di-undo"

# Contoh penggunaan
npm_last_two_digits = 23  # Ganti dengan dua digit terakhir NPM Anda
calc = Calculator(npm_last_two_digits)

print("Initial Value:", calc.stack[-1])
print("Add 5:", calc.add(5))
print("Subtract 3:", calc.subtract(3))
print("Multiply by 2:", calc.multiply(2))
print("Undo:", calc.undo())
print("Undo:", calc.undo())

```

```python
from collections import deque

class MobileApp:
    def __init__(self):
        self.history = deque()

    def visit_feature(self, feature):
        self.history.append(feature)
        print(f"Mengunjungi fitur: {feature}")

    def back(self):
        if self.history:
            print(f"Kembali ke: {self.history.pop()}")
        else:
            print("Tidak ada riwayat!")

# Contoh penggunaan
birth_month = 'J'  # Ganti dengan huruf pertama bulan lahir Anda
npm_last_two_digits = '23'  # Ganti dengan dua digit terakhir NPM Anda

app = MobileApp()
app.visit_feature(f"{birth_month}{npm_last_two_digits}_Fitur1")
app.visit_feature(f"{birth_month}{npm_last_two_digits}_Fitur2")
app.visit_feature(f"{birth_month}{npm_last_two_digits}_Fitur3")

app.back()
app.back()
```

```python
from queue import LifoQueue

class EmailDraftSystem:
    def __init__(self):
        self.drafts = LifoQueue()

    def save_draft(self, subject):
        self.drafts.put(subject)
        print(f"Menambahkan draft: {subject}")

    def view_latest_draft(self):
        if not self.drafts.empty():
            return f"Draft terakhir: {self.drafts.queue[-1]}"
        return "Tidak ada draft!"

# Contoh penggunaan
name_initials = 'Ol'  # Ganti dengan dua huruf pertama nama Anda
npm_last_two_digits = '23'  # Ganti dengan dua digit terakhir NPM Anda

email_system = EmailDraftSystem()
email_system.save_draft(f"{name_initials}{npm_last_two_digits}_Draft1")
email_system.save_draft(f"{name_initials}{npm_last_two_digits}_Draft2")
email_system.save_draft(f"{name_initials}{npm_last_two_digits}_Draft3")

print(email_system.view_latest_draft())
```

```python
class TowerOfHanoi:
    def __init__(self, num_disks):
        self.num_disks = num_disks
        self.towers = [list(range(num_disks, 0, -1)), [], []]
        self.max_moves = 23  # Ganti dengan dua digit terakhir NPM Anda

    def move_disk(self, from_tower, to_tower):
        if self.towers[from_tower]:
            disk = self.towers[from_tower].pop()
            self.towers[to_tower].append(disk)
            print(f"Memindahkan disk {disk} dari menara {from_tower + 1} ke menara {to_tower + 1}")
        else:
            print("Tidak ada disk yang bisa dipindahkan.")

    def solve(self, n, source, target, auxiliary):
        if n == 0:
            return
        self.solve(n - 1, source, auxiliary, target)
        self.move_disk(source, target)
        self.solve(n - 1, auxiliary, target, source)

# Contoh penggunaan
hanoi = TowerOfHanoi(3)
hanoi.solve(hanoi.num_disks, 0, 2, 1)
```

IMPLEMENTASI STACK PADA PENUMPUKAN BARANG DI GUDANG

```python
class Menu:
    def __init__(self, max_items):
        self.stack = []  # Stack untuk menyimpan barang
        self.max_items = max_items  # Kapasitas maksimal barang dalam gudang

    def add_item(self, item):
        """Menambahkan barang ke dalam gudang (Stack)."""
        if len(self.stack) < self.max_items:
            self.stack.append(item)
            print(f"Menambahkan barang: {item}")
        else:
            print("Gudang penuh! Tidak bisa menambahkan barang lagi.")

    def hapus_barang(self):
        """Menghapus barang teratas dari gudang (Stack)."""
        if self.stack:
            item = self.stack.pop()
            print(f"Mengambil barang: {item}")
        else:
            print("Gudang kosong! Tidak ada barang yang bisa diambil.")

    def daftar_barang(self):
        """Menampilkan daftar barang yang ada di gudang."""
        if self.stack:
            print("Daftar barang di gudang:")
            for i, item in enumerate(self.stack, start=1):
                print(f"{i}. {item}")
        else:
            print("Gudang kosong.")

# Kapasitas maksimal barang dalam gudang
max_items = 23
menu = Menu(max_items)

# Menu interaktif
while True:
    print("\nMenu:")
    print("1. Tambahkan barang ke gudang")
    print("2. Ambil barang dari gudang")
    print("3. Lihat daftar barang di gudang")
    print("4. Keluar")

    X = input("Pilih menu (1/2/3/4): ")

    if X == "1":
        item = input("Masukkan nama barang: ")
        menu.add_item(item)
    elif X == "2":
        remove_iter = input("Masukkan jumlah barang yang ingin diambil: ")
        if remove_iter.isdigit():
            remove_iter = int(remove_iter)
            if remove_iter <= len(menu.stack):
                for _ in range(remove_iter):
                    menu.hapus_barang()
            else:
                print("Jumlah barang yang ingin diambil melebihi stok gudang.")
        else:
            print("Masukkan angka yang valid.")
    elif X == "3":
        menu.daftar_barang()
    elif X == "4":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid. Silakan pilih menu yang benar (1/2/3/4).")
```
