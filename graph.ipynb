# Materi


 Representasi Graf dengan Adjacency List

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['A','F'],
    'D': ['B'],
    'E': ['B','F'],
    'F': ['C','E']
}

def print_garph(graf):
  for node in graf:
    print(f"{node} -> {graf[node]}")

print_garph(graph)
```

Taversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
  visited = set ()
  queue = deque ([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])
bfs(graph,'A')
```

Tranversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set ()

  if start not in visited:
    print(start, end="")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

dfs(graph,'A')
```

Latihan Personal

```python
def count_neighbors(graph) :
  result = {}
  for node in graph :
    result[node] = len(graph[node])
  return result

print(count_neighbors(graph))


```

# LATIHAN / TUGAS

**1. Membuat Graf Tak Berarah (5 Simpul & 7 Sisi)**

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'E'],
    'D': ['B', 'E', 'F'],
    'E': ['B', 'C', 'D', 'F'],
    'F': ['D', 'E']
}

def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

print_graph(graph)
```

**2. Implementasi BFS dan DFS**

BFS (Mulai dari simpul 'A'):

```python
from collections import deque

def bfs(graph, start):
    visited = []
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.append(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])
    return visited

print("BFS:", bfs(graph, 'A'))
```

**DFS (Mulai dari simpul 'A'):**

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = []
    if start not in visited:
        visited.append(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)
    return visited

print("DFS:", dfs(graph, 'A'))
```

Fungsi find_path (Menggunakan DFS)

```python
def find_path(graph, start, end, path=None):
    if path is None:
        path = []
    path = path + [start]
    if start == end:
        return path
    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path:
                return new_path
    return None

print("Path dari A ke F:", find_path(graph, 'A', 'F'))
```

3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan, bukan hanya
mencetak.


```python
from collections import deque


# Variabel
test_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Fungsi
def bfs(graph, start):
  visit = []
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visit.append(vertex)
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

  return visit


# Program Utama
print("BFS:")
print(f"\n\nUrutan:\n{bfs(test_graph, 'A')}")
```

4. fungsi find_path(graph, start, end) yang mengembalikan satu jalur dari start
ke end jika ada, menggunakan DFS.


```python
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path.append(start)

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in visited:
            result = find_path(graph, neighbor, end, visited, path)
            if result:
                return result

print("Jalur dari A ke D:")
find_path(test_graph, 'A', 'D')
```

5. Tambahkan fungsi is_connected(graph) untuk mengecek apakah semua simpul
dalam graf saling terhubung (gunakan kombinasi DFS + logika).

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

    return visited

def is_connected(graph):
    # Ambil simpul awal secara acak
    start_node = next(iter(graph))
    visited = dfs(graph, start_node)

    # Cek apakah semua simpul sudah dikunjungi
    return len(visited) == len(graph)

is_connected(test_graph)
print("Apakah graf terhubung?", is_connected(test_graph))
```
